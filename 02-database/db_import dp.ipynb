## Steps
1. Extract skills
2. Extract education level
3. Store into database
4. Export the csv for backup

## Instructions

* Edit the connection info in the next cell
* Can extract more information from job description by using the same pattern

```python
mysql_address='localhost'
mysql_user='root'
mysql_pwd='root'
db_name='jobsdbtest'
table_name='computer_vision_jobs'
```

```python
import pandas as pd
```

```python
df=pd.read_csv('../00-data/joblist_combined_bk.csv')
df['skills']=""
df['education']=""
#df['subject']=""
df.head()
```

```python
skill_list = ['sql', 'nosql', 'python', 'java', 'c++', 'ruby', 'r', 'julia', 'perl', 'matlab', 'mathematica', 'php', 'javascript', 'scala', 'octave', 'c/c++', 'pytorch', 'mongodb', 'mysql','js','hbase', 'cassandra', 'spark','mapreduce', 'tableau', 'sas', 'spss', 'cnn', 'kcnn', 'dcnns']
degree_list = ['bachelor', 'master', 'phd', 'bs', 'ms']
```

```python
def extract_skills(desc_str):
    skill_ext_set=set()
    description_text = str(desc_str).lower()
    for skill in skill_list:
        if skill in description_text:
            skill_ext_set.add(skill)
    
    skill_set = ','.join(skill_ext_set)
    return skill_set
    
def extract_degree(desc_str):
    skill_ext_set=set()
    description_text = str(desc_str).lower()
    for degree in degree_list:
        if degree in description_text:
            skill_ext_set.add(degree)
    
    skill_set = ','.join(skill_ext_set)
    return skill_set
    #row['skills']=skill_set
```

```python
df.dtypes
```

```python
df['skills']=df['description'].apply(extract_skills)
df['education']=df['description'].apply(extract_degree)
df.head()
```

```python
from pandas.io import sql
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://{user}:{pw}@{address}/{db}"
                       .format(user=mysql_user,
                               pw=mysql_pwd,
                               db=db_name, address=mysql_address))
df.to_sql(con=engine, name=table_name, if_exists='replace')
```

```python
df.to_csv('../00-data/jobdb_bk.csv',encoding='utf-8', index=False)
```
